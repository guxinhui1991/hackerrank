```python
def number_base_arithmetic(base, number):
    if(int(base)<2 or int(base)>36):
        return 'base invalid'
    
    base_vals = {}
    for i in range(0, int(base)):
        if(i < 10):
            base_vals[str(i)] = i
        else:
            base_vals[str(chr(ord('a')+ i - 10))] = i
            
    for i in range(len(number)):
        if(number[i] not in base_vals.keys()):
            return 'number invalid'
            break
    
    val_base10 = 0
    for i in range(len(number)):
        val_base10 = val_base10 + base_vals[number[len(number)-i-1]]*int(base)**i
    return str(val_base10)
```

```python
def find_words_in_graph(list_of_words, edge_list):
    
    def search_graph(v, edge_list):
        # Base case:
        if len(v)==1:
            return 1
                
        # Recursive case:  
        else:
            for i in range(len(v)-1):
                if not search_graph(v[i:i+1], edge_list):
                    return 0
            return 1

    res = [0]*len(list_of_words)
    for i, w in enumerate(list_of_words):
        res[i] = search_graph(w, edge_list)

    return res
      
```

```python
def isBipartite(self, graph):
    """
    :type graph: List[List[int]]
    :rtype: bool
    """
    
    # Use BFS to check path between s and d
    def isReachable(self, s, d):
        # Mark all the vertices as not visited
        visited =[False]*(self.V)
  
        # Create a queue for BFS
        queue=[]
  
        # Mark the source node as visited and enqueue it
        queue.append(s)
        visited[s] = True
  
        while queue:
 
            #Dequeue a vertex from queue 
            n = queue.pop(0)
             
            # If this adjacent node is the destination node,
            # then return true
             if n == d:
                 return True
 
            #  Else, continue to do BFS
            for i in self.graph[n]:
                if visited[i] == False:
                    queue.append(i)
                    visited[i] = True
         # If BFS is complete without visited d
         return False
```
